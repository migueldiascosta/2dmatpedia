```python
import json
import sys
import os

from pymatgen.electronic_structure.bandstructure import BandStructureSymmLine
from pymatgen.electronic_structure.dos import CompleteDos
from pymatgen.electronic_structure.plotter import BSDOSPlotter
```

```python
filename_bands='2dm-152.bands.json'
filename_dos='2dm-152.dos.json'

bands_dict=json.load(open(filename_bands))
dos_dict=json.load(open(filename_dos))
```

```python
bands=BandStructureSymmLine.from_dict(bands_dict)
dos=CompleteDos.from_dict(dos_dict)
```

```python
%matplotlib inline
bsp=BSDOSPlotter()
bsp.get_plot(bands, dos=dos)
```
